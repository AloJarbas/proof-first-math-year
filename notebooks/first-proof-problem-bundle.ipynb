# First proof problem bundle

A compact companion notebook for the opening proof-first block.
The point is not volume. The point is to build control over definitions, quantifiers, proof form, and self-checking.


![Proof-first timeline](../assets/proof-first-math-year-timeline.png)

This notebook belongs to the very beginning of that timeline: before curriculum sprawl, before four parallel courses, while proof fluency is still the main bottleneck.


## Scope and boundary

**Question:** what should an opening proof bundle reinforce so the learner becomes better at *reading, selecting, and checking* proof moves rather than just accumulating pages?

**In scope:** statement anatomy, definition use, counterexample instinct, direct proof structure, set language, and light induction preview.

**Out of scope:** heavy algebra drills, long olympiad-style ingenuity problems, and broad topic coverage for its own sake.


## Core concepts to revisit before attempting the bundle

1. **Conditional form**: if \(P\), then \(Q\).
2. **Universal claim**: \(\forall x \in S,\; P(x)\).
3. **Existence claim**: \(\exists x \in S\) such that \(P(x)\).
4. **Set inclusion**: to prove \(A \subseteq B\), start with an arbitrary \(x \in A\) and show \(x \in B\).
5. **Counterexample logic**: a single valid counterexample kills a false universal statement.

A useful self-check question is: *which definition actually has to appear in the proof for it to be real?*


## Problem bundle

### Problem 1 — Statement anatomy
Rewrite the claim in your own words and identify hypotheses, conclusion, and hidden quantifiers:

> If an integer \(n\) is divisible by 6, then \(n\) is divisible by 3.

**Check:** where does divisibility need to be unpacked formally?

### Problem 2 — Counterexample instinct
Find a counterexample to:

> For every real number \(x\), if \(x^2 > x\), then \(x > 1\).

**Check:** why does one example suffice here?

### Problem 3 — Direct proof
Prove:

> If \(a\) and \(b\) are odd integers, then \(a+b\) is even.

**Check:** which representation of odd numbers makes the proof move obvious?

### Problem 4 — Set inclusion
Let \(A, B, C\) be sets. Prove that if \(A \subseteq B\) and \(B \subseteq C\), then \(A \subseteq C\).

**Check:** are you proving subset inclusion directly, or talking vaguely *about* subsets?

### Problem 5 — Contrapositive choice
Prove:

> If \(n^2\) is even, then \(n\) is even.

**Check:** is direct proof or contrapositive cleaner here, and why?

### Problem 6 — Induction preview
Prove for every integer \(n \ge 1\):

\[1 + 2 + \cdots + n = \frac{n(n+1)}{2}.\]

**Check:** separate the base case, induction hypothesis, and induction step cleanly.


## Worked-example audit prompts

When you look at a solved proof, do not only ask whether it seems correct. Ask:

- Why was this proof form chosen?
- Which definition did the real work?
- Which line would fail first if a hypothesis changed?
- Can I hide the last third and finish it myself?

That is how a worked example becomes reusable instead of decorative.


## Adversarial check

The obvious failure mode for a first proof bundle is **fake variety**: mixing too many problem types too early and mistaking confusion for rigor.

So this bundle stays small and method-centered. The goal is not to cover everything. The goal is to make the learner better at recognizing:

- when a definition must be unpacked,
- when a counterexample is enough,
- when direct proof is natural,
- when contrapositive is cleaner,
- how induction structure is actually laid out.
