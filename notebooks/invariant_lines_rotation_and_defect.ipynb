## Case 1: diagonal scaling

Let

\[
D = \begin{pmatrix} 3 & 0 \\ 0 & 1/2 \end{pmatrix}.
\]

1. Show that `\operatorname{span}\{(1,0)\}` is invariant.
2. Show that `\operatorname{span}\{(0,1)\}` is invariant.
3. Write one sentence explaining why those two lines already explain the diagonal form.

**Check:** the plane already splits into two independent closed lanes.


## Case 2: real rotation

Let

\[
R = \begin{pmatrix} \cos(\pi/3) & -\sin(\pi/3) \\ \sin(\pi/3) & \cos(\pi/3) \end{pmatrix}.
\]

1. Explain why the whole plane is still invariant.
2. Explain why a nonzero real line through the origin cannot stay invariant under this rotation.
3. Write one sentence separating "the plane is invariant" from "there is a real eigenline."

**Check:** this failure is not "not enough lines." It is "no real line survives."


## Case 3: defective Jordan block

Let

\[
J = \begin{pmatrix} 2 & 1 \\ 0 & 2 \end{pmatrix}.
\]

1. Show that the x-axis is invariant.
2. Check directly that the y-axis is not invariant.
3. Write one sentence explaining why one surviving eigenline does not yet give a full diagonal split.

**Check:** this failure is not "no line survives." It is "one line survives, but the second lane never closes."
