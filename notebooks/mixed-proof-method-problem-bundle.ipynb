# Mixed proof-method problem bundle

This companion bundle is for the moment when a learner can already write short proofs, but still misclassifies the problem in front of them.

The point is not to make the problems harder by default. The point is to make **method selection, self-explanation, and retrospective checking** impossible to skip.


## Scope and boundary

**Question:** what should a second proof bundle train once the first notebook has already covered definitions, direct proof, counterexamples, and an induction preview?

**In scope:** mixed method choice, contrapositive versus contradiction, short case splits, induction with a cleaner checkpoint, and active reading of worked proof structure.

**Out of scope:** olympiad-style ingenuity, long algebra manipulations, and topic sprawl disguised as rigor.


## Why this bundle is different from the first one

The first proof notebook was about getting proof form on the table at all.
This one is about the next failure mode: writing something that looks like a proof before deciding which method actually fits.

So every problem in this bundle starts with the same three checks:

1. **Method pick:** direct, contrapositive, contradiction, cases, or induction?
2. **Why this method first?** One sentence only.
3. **What would the wrong first move probably look like?**


## Warm-up: method-choice drill

Before solving anything, classify these quickly:

- If `n^2` is odd, then `n` is odd.
- There is no smallest positive rational number.
- Every integer is either even or odd.
- For all `n >= 1`, `1 + 3 + ⋯ + (2n-1) = n^2`.

Then check which classifications you would defend and which ones only felt familiar.


## Problem bundle

### Problem 1 — Contrapositive clean-up
Prove:

> If an integer `n` is not divisible by 3, then `n^2` is not divisible by 3.

**Method pick:** which is cleaner here: direct proof or contrapositive?

**Answer-check scaffold:** if you use the contrapositive, can you prove the reverse statement by reducing the possibilities for `n mod 3`?

### Problem 2 — Contradiction versus direct proof
Prove that there is no integer `n` such that `n` and `n+1` are both even.

**Method pick:** contradiction or direct unpacking?

**Answer-check scaffold:** where exactly does the parity clash appear?

### Problem 3 — Proof by cases without bookkeeping bloat
Prove that for every integer `n`, the number `n(n+1)` is even.

**Method pick:** would you rather use cases or divisibility structure?

**Answer-check scaffold:** if you split into even/odd cases, does each case actually finish in two or three lines? If not, you are probably over-writing it.

### Problem 4 — Set proof with a hidden method choice
Let `A`, `B`, and `C` be sets. Prove that

`A ∩ (B ∪ C) = (A ∩ B) ∪ (A ∩ C)`.

**Method pick:** are you proving equality by double inclusion, or drifting into element-chasing without a clear target?

**Answer-check scaffold:** can you write the left-to-right inclusion and the right-to-left inclusion with the same skeleton?

### Problem 5 — Counterexample first, proof second
Decide whether the statement is true or false:

> For all real numbers `x` and `y`, if `xy = 1`, then `x = y`.

If it is false, give the smallest clean counterexample you can and say why one example is enough.

**Answer-check scaffold:** do not repair a false statement into a true one before you have killed the original claim honestly.

### Problem 6 — Induction with a real step
Prove for every integer `n >= 1` that

`1 + 2 + ⋯ + n = n(n+1)/2`.

Then write one extra sentence:

- what exactly changed from `n` to `n+1`,
- and where the induction hypothesis was actually used.

**Answer-check scaffold:** if the induction hypothesis disappears from your step, the proof is fake.

### Problem 7 — A slightly harder method-choice problem
Prove:

> If `a^2 + b^2` is even for integers `a` and `b`, then `a` and `b` have the same parity.

**Method pick:** direct proof, contrapositive, or cases?

**Answer-check scaffold:** can you classify the parity pairs `(even, odd)` and `(odd, even)` quickly enough to see what the cleanest route should be?


## Retrospective audit after each proof

Do not stop at 'correct' or 'incorrect'.
After each problem, log these:

- Which method I chose first
- Which method would have been cleaner in retrospect
- Which definition or fact did the real work
- Which line would collapse first if I had to reproduce the proof tomorrow with no notes


## Worked-example self-explanation prompts

For one solved proof from class notes or a textbook, pause after each line and ask:

- What idea is this line using?
- Why was that idea used here instead of a different one?
- Which previous theorem, definition, or small fact is being imported silently?
- If I hide the next two lines, can I continue the argument honestly?

That turns worked examples into training material instead of decoration.


## Adversarial check

The failure mode here is easy to fake: a mixed bundle can look deep while really being just a random pile of proofs.

This notebook tries to block that in three ways:

1. every problem has a method-choice checkpoint,
2. every problem has an answer-check scaffold,
3. the bundle stays on short arguments where method recognition matters more than grind.

If the learner is still misclassifying the problem type, adding harder theorems too early will only hide the real bottleneck.


## Source basis and intake mix

This pass used **Raindrop + HN + source docs/pages** as the intake mix.

### Accepted

1. **The Learning Scientists — Worked Examples: An Effective Tool for Math Learning**
   Accepted for the case that worked examples help novices most when paired with active self-explanation rather than passive rereading.

2. **Lancaster / Loughborough self-explanation guide for proofs**
   Accepted because it gives a concrete line-by-line proof-reading habit instead of vague advice to 'understand the proof better.'

3. **Existing repo notes**
   Accepted because this bundle is supposed to deepen the repo's own proof-method and weekly-review packet, not replace it with imported pedagogy slogans.

### Rejected

1. **HN discussion on learning proofs from scratch**
   Rejected as a primary spine. It was useful for quick triangulation, but not strong enough to carry the bundle design.

2. **Generic proof-writing handouts**
   Rejected when they focused on formatting and politeness more than method choice or self-checking.
