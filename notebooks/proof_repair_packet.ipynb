## Subset chains still need a first element

**Source problem:** Subset chain  
**Primary step family:** choose an arbitrary object  
**Method:** direct  
**Prompt:** Given A ⊆ B and B ⊆ C, prove A ⊆ C.

### The weak line
> Since B ⊆ C, anything in B is in C.

### Why it fails
That sentence starts in the middle. The proof never chooses an arbitrary x ∈ A, so the inclusion A ⊆ B never gets used on the same object.

### Repair cue
Restart from one arbitrary x in A, then push that exact x through both inclusions.


### Working page

**Answer-check prompts from the source problem**
- Did you start with an arbitrary x ∈ A instead of a special element?
- Did you actually use both subset hypotheses?
- Did you end with “therefore A ⊆ C”, not just “x ∈ C”?

**Repaired hinge**
- Let x be arbitrary with x ∈ A. Since A ⊆ B, we have x ∈ B; since B ⊆ C, we have x ∈ C. Therefore A ⊆ C.

- If the claim were A ∩ B ⊆ A, what is the first arbitrary object you would choose?

## Divisibility needs an actual witness

**Source problem:** Divisibility under addition  
**Primary step family:** name the hidden witness or coefficients  
**Method:** direct  
**Prompt:** If a | b and a | c, prove a | (b + c).

### The weak line
> Both b and c have a factor a, so b + c does too.

### Why it fails
That is only a slogan. The integers witnessing a | b and a | c are still hidden, so the new witness for b + c never appears on the page.

### Repair cue
Name b = am and c = an first, then use m + n as the new witness instead of waving at a common factor.


### Working page

- Starter line from the source problem: Write b = ________ and c = ________. Then b + c = ________.
- Your diagnosis in one sentence:
- Your rewrite before looking at the repaired hinge:
- Which word or phrase was doing fake work in the bad line?

**Answer-check prompts from the source problem**
- Did you introduce two integers witnessing a | b and a | c?
- Did you factor out a after adding the two expansions?
- Did you identify the new witness, rather than just circling the factor a?

**Repaired hinge**
- Since a | b and a | c, there exist integers m and n with b = am and c = an. Then b + c = a(m + n), so a | (b + c).

**Transfer check**
- If the claim were a | (b − c), which witness changes and which part of the proof stays the same?

**One more rewrite**

## A subspace proof cannot hide behind “by linearity”

**Source problem:** Kernel is a subspace  
**Primary step family:** split the closure or case step cleanly  
**Method:** subspace proof  
**Prompt:** Let T: V → W be linear. Prove that Ker(T) is a subspace of V.

### The weak line
> Ker(T) is closed under addition and scalar multiplication by linearity.

### Why it fails
That skips the whole structure. A kernel subspace proof still owes three visible checkpoints: zero vector, additive closure, and scalar closure.

### Repair cue
Write the three checkpoints separately and compute T(u + v) and T(αu) instead of summarizing them with one vague sentence.


### Working page

- Starter line from the source problem: Show three things: 0 ∈ Ker(T), then if u, v ∈ Ker(T) prove ________, and if α is a scalar prove ________.
- Your diagnosis in one sentence:
- Your rewrite before looking at the repaired hinge:
- Which word or phrase was doing fake work in the bad line?

**Answer-check prompts from the source problem**
- Did you verify the zero vector separately?
- Did you compute T(u + v) and T(αu) using linearity, not just say “closure holds”?
- Did you conclude that all three subspace conditions are satisfied?

**Repaired hinge**
- First T(0) = 0, so 0 ∈ Ker(T). If u, v ∈ Ker(T), then T(u + v) = T(u) + T(v) = 0, so u + v ∈ Ker(T). If α is a scalar and u ∈ Ker(T), then T(αu) = αT(u) = 0, so αu ∈ Ker(T). Therefore Ker(T) is a subspace of V.

**Transfer check**
- If you were proving Im(T) is a subspace, which checkpoint would need a new witness and which two would stay the same?

## A picture is not the contradiction

**Source problem:** A real rotation has no nonzero invariant line  
**Primary step family:** finish the algebra or coordinate cleanup  
**Method:** contradiction / components  
**Prompt:** Let R be rotation by 60° in ℝ². Prove that there is no nonzero real vector u with Ru = λu for some real λ.

### The weak line
> A 60° rotation changes the direction, so no nonzero real line can stay invariant.

### Why it fails
The intuition is fine, but the proof never turns “same line” into Ru = λu and never squeezes a contradiction out of the coordinate equations.

### Repair cue
Take u = (x, y) ≠ 0, assume Ru = λu for some real λ, and force the two rotated coordinates to agree with one real scalar.


### Working page

- Starter line from the source problem: Take a nonzero u = (x, y) and assume Ru = λu. Compare the two coordinates of Ru with λx and λy.
- Your diagnosis in one sentence:
- Your rewrite before looking at the repaired hinge:
- Which word or phrase was doing fake work in the bad line?

**Answer-check prompts from the source problem**
- Did you start from one nonzero vector u and one real scalar λ?
- Did you compare both coordinates instead of arguing from a picture alone?
- Did the contradiction rule out every nonzero real invariant line, not just one sample vector?

**Repaired hinge**
- Let u = (x, y) ≠ 0 and assume Ru = λu with λ ∈ ℝ. Then (x/2 − √3 y/2, √3 x/2 + y/2) = (λx, λy). From the two coordinates we get √3 y = (1 − 2λ)x and √3 x = (2λ − 1)y. Multiplying gives 3xy = −(2λ − 1)^2 xy, impossible when xy ≠ 0; if x = 0 or y = 0, the equations force the other coordinate to vanish too. Contradiction. Therefore no nonzero real invariant line exists.

**Transfer check**
- For a 90° rotation, which real-scalar equation would fail first if you tried the same Ru = λu setup?

## Old independence cannot be used too early

**Source problem:** Adjoining a vector outside the span  
**Primary step family:** turn the false branch into a contradiction  
**Method:** contradiction / linear combination  
**Prompt:** If v₁, …, v_k are linearly independent and v ∉ span(v₁, …, v_k), prove that v₁, …, v_k, v are linearly independent.

### The weak line
> Because v₁, …, v_k are linearly independent, all coefficients are zero.

### Why it fails
Not yet. The old independence only applies after the new coefficient β has been forced to zero. Until then, the relation still contains the extra vector v.

### Repair cue
Start from one generic linear relation, isolate the β ≠ 0 branch, and use it to push v back into the old span before invoking the old independence.


### Working page

- Starter line from the source problem: Assume α₁v₁ + ··· + α_kv_k + βv = 0. What happens if β ≠ 0?
- Your diagnosis in one sentence:
- Your rewrite before looking at the repaired hinge:
- Which word or phrase was doing fake work in the bad line?

**Answer-check prompts from the source problem**
- Did you start from one generic linear relation equal to zero?
- Did you split the proof into the β ≠ 0 and β = 0 possibilities implicitly or explicitly?
- Did you use the original independence only after forcing β = 0?

**Repaired hinge**
- Assume α₁v₁ + ··· + α_kv_k + βv = 0. If β ≠ 0, then v = −(α₁/β)v₁ − ··· − (α_k/β)v_k, so v lies in span(v₁, …, v_k), contradiction. Hence β = 0. Then α₁v₁ + ··· + α_kv_k = 0, so the old independence forces α₁ = ··· = α_k = 0. Therefore v₁, …, v_k, v are linearly independent.

**Transfer check**
- In a basis-extension proof, which coefficient has to be isolated before the old independence becomes legal to use?

## The computation is not the full inclusion

**Source problem:** T² = 0 forces Im(T) ⊆ Ker(T)  
**Primary step family:** say exactly what was proved  
**Method:** direct  
**Prompt:** Let T: V → V be linear and suppose T² = 0. Prove Im(T) ⊆ Ker(T).

### The weak line
> So T(y) = 0.

### Why it fails
That line only computes the kernel condition. The proof still has to say y ∈ Ker(T) and then cash the arbitrary choice of y out into Im(T) ⊆ Ker(T).

### Repair cue
After the computation, restate the membership fact and close the arbitrary-element argument as an inclusion sentence.


### Working page

- Starter line from the source problem: Take y ∈ Im(T). Then y = ________ for some x ∈ V. Now compute T(y) = ________.
- Your diagnosis in one sentence:
- Your rewrite before looking at the repaired hinge:
- Which word or phrase was doing fake work in the bad line?

**Answer-check prompts from the source problem**
- Did you choose y from Im(T), not from all of V?
- Did you name a preimage x with y = T(x)?
- Did you end by saying y ∈ Ker(T), so Im(T) ⊆ Ker(T)?

**Repaired hinge**
- Take y ∈ Im(T). Then y = T(x) for some x ∈ V. Hence T(y) = T(T(x)) = T²(x) = 0, so y ∈ Ker(T). Because y was arbitrary in Im(T), we conclude Im(T) ⊆ Ker(T).

**Transfer check**
- If you prove every y ∈ Y lies in Z, what closing sentence turns that pointwise fact into Y ⊆ Z?
