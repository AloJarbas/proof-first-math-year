## How to use the notebook

1. Read one prompt.
2. Write a proof attempt before checking the scaffold prompts.
3. Use the answer-check bullets to find the first vague sentence.
4. Rewrite that sentence so the hidden move is explicit.


## Row 1: Subset chain

**Stage:** proof fluency  
**Method:** direct  
**Difficulty:** 1

**Prompt:** Given A ⊆ B and B ⊆ C, prove A ⊆ C.

**Starter line:** Let x be ________. Then x ∈ B because ________, and x ∈ C because ________.

**Highlighted missing moves**

- choose an arbitrary object

**Answer-check prompts**

- Did you start with an arbitrary x ∈ A instead of a special element?
- Did you actually use both subset hypotheses?
- Did you end with “therefore A ⊆ C”, not just “x ∈ C”?

**Common slip:** Starting from x ∈ B or x ∈ C skips the only structural move that matters.

**Reflection prompt:** Which sentence in your proof carried the actual structural load, and did you write it explicitly or only imply it?


## Row 2: Divisibility under addition

**Stage:** proof fluency  
**Method:** direct  
**Difficulty:** 2

**Prompt:** If a | b and a | c, prove a | (b + c).

**Starter line:** Write b = ________ and c = ________. Then b + c = ________.

**Highlighted missing moves**

- name the hidden witness or coefficients
- finish the algebra or coordinate cleanup

**Answer-check prompts**

- Did you introduce two integers witnessing a | b and a | c?
- Did you factor out a after adding the two expansions?
- Did you identify the new witness, rather than just circling the factor a?

**Common slip:** Saying “both have a factor a” is not yet a proof unless one witness for b + c is written down.

**Reflection prompt:** Which sentence in your proof carried the actual structural load, and did you write it explicitly or only imply it?


## Row 3: Kernel is a subspace

**Stage:** linear algebra bridge  
**Method:** subspace proof  
**Difficulty:** 3

**Prompt:** Let T: V → W be linear. Prove that Ker(T) is a subspace of V.

**Starter line:** Show three things: 0 ∈ Ker(T), then if u, v ∈ Ker(T) prove ________, and if α is a scalar prove ________.

**Highlighted missing moves**

- choose an arbitrary object
- split the closure or case step cleanly
- say exactly what was proved

**Answer-check prompts**

- Did you verify the zero vector separately?
- Did you compute T(u + v) and T(αu) using linearity, not just say “closure holds”?
- Did you conclude that all three subspace conditions are satisfied?

**Common slip:** Writing “kernels are always subspaces” is a memory cue, not the proof.

**Reflection prompt:** Which sentence in your proof carried the actual structural load, and did you write it explicitly or only imply it?


## Row 4: T² = 0 forces Im(T) ⊆ Ker(T)

**Stage:** linear algebra bridge  
**Method:** direct  
**Difficulty:** 3

**Prompt:** Let T: V → V be linear and suppose T² = 0. Prove Im(T) ⊆ Ker(T).

**Starter line:** Take y ∈ Im(T). Then y = ________ for some x ∈ V. Now compute T(y) = ________.

**Highlighted missing moves**

- choose an arbitrary object
- name the hidden witness or coefficients
- say exactly what was proved

**Answer-check prompts**

- Did you choose y from Im(T), not from all of V?
- Did you name a preimage x with y = T(x)?
- Did you end by saying y ∈ Ker(T), so Im(T) ⊆ Ker(T)?

**Common slip:** If the hidden preimage never appears, the proof cannot use the hypothesis T² = 0 in the right place.

**Reflection prompt:** Which sentence in your proof carried the actual structural load, and did you write it explicitly or only imply it?


## Row 5: Adjoining a vector outside the span

**Stage:** linear algebra bridge  
**Method:** contradiction / linear combination  
**Difficulty:** 4

**Prompt:** If v₁, …, v_k are linearly independent and v ∉ span(v₁, …, v_k), prove that v₁, …, v_k, v are linearly independent.

**Starter line:** Assume α₁v₁ + ··· + α_kv_k + βv = 0. What happens if β ≠ 0?

**Highlighted missing moves**

- name the hidden witness or coefficients
- turn the false branch into a contradiction
- say exactly what was proved

**Answer-check prompts**

- Did you start from one generic linear relation equal to zero?
- Did you split the proof into the β ≠ 0 and β = 0 possibilities implicitly or explicitly?
- Did you use the original independence only after forcing β = 0?

**Common slip:** Using the old independence too early misses the actual reason the new vector is safe to adjoin.

**Reflection prompt:** Which sentence in your proof carried the actual structural load, and did you write it explicitly or only imply it?


## Row 6: A real rotation has no nonzero invariant line

**Stage:** geometry follow-up  
**Method:** contradiction / components  
**Difficulty:** 5

**Prompt:** Let R be rotation by 60° in ℝ². Prove that there is no nonzero real vector u with Ru = λu for some real λ.

**Starter line:** Take a nonzero u = (x, y) and assume Ru = λu. Compare the two coordinates of Ru with λx and λy.

**Highlighted missing moves**

- choose an arbitrary object
- finish the algebra or coordinate cleanup
- turn the false branch into a contradiction
- say exactly what was proved

**Answer-check prompts**

- Did you start from one nonzero vector u and one real scalar λ?
- Did you compare both coordinates instead of arguing from a picture alone?
- Did the contradiction rule out every nonzero real invariant line, not just one sample vector?

**Common slip:** Pointing at a picture of a rotated arrow is good intuition, but it is not yet the contradiction.

**Reflection prompt:** Which sentence in your proof carried the actual structural load, and did you write it explicitly or only imply it?


## One useful meta-question

Which missing move keeps failing across multiple rows? If the same move breaks twice, fix that move before asking for harder problems.
