## Problem 1 — subspace proof instead of shape intuition

Let

\[
S = \{(x,y,z) \in \mathbb{R}^3 : x - 2y + z = 0\}.
\]

### Prompt

1. Prove that `S` is a subspace of `\mathbb{R}^3`.
2. Produce a basis for `S`.
3. State its dimension.

### Worked answer

To prove `S` is a subspace, check the three standard items.

**Zero vector:** `(0,0,0)` lies in `S` because `0 - 2\cdot 0 + 0 = 0`.

**Closed under addition:** if `(x_1,y_1,z_1)` and `(x_2,y_2,z_2)` lie in `S`, then

\[
x_1 - 2y_1 + z_1 = 0, \qquad x_2 - 2y_2 + z_2 = 0.
\]

Add the equations:

\[
(x_1+x_2) - 2(y_1+y_2) + (z_1+z_2) = 0.
\]

So `(x_1+x_2, y_1+y_2, z_1+z_2)` also lies in `S`.

**Closed under scalar multiplication:** if `(x,y,z)` lies in `S` and `c \in \mathbb{R}`, then

\[
cx - 2(cy) + cz = c(x - 2y + z) = c \cdot 0 = 0.
\]

So `(cx,cy,cz)` also lies in `S`. Therefore `S` is a subspace.

Now solve for one coordinate:

\[
z = -x + 2y.
\]

So every vector in `S` has the form

\[
(x,y,-x+2y) = x(1,0,-1) + y(0,1,2).
\]

Hence

\[
S = \operatorname{span}\{(1,0,-1),(0,1,2)\}.
\]

These two vectors are independent because

\[
a(1,0,-1) + b(0,1,2) = (0,0,0)
\]

forces `a=0` from the first coordinate and `b=0` from the second.
So a basis is

\[
\bigl\{(1,0,-1),(0,1,2)\bigr\},
\]

and the dimension is `2`.

### Structural sentence

One linear equation cuts one degree of freedom out of `\mathbb{R}^3`, so this set is a plane through the origin, not just a random slice that happens to look flat.

### Likely wrong move

Saying 'it looks like a plane, so it must be a subspace' skips the actual closure argument and would fail for affine planes that do **not** pass through the origin.


## Problem 2 — kernel and image as proof objects

Define

\[
T : \mathbb{R}^3 \to \mathbb{R}^2, \qquad T(x,y,z) = (x+y, y+z).
\]

### Prompt

1. Find `\ker(T)`.
2. Find `\operatorname{im}(T)`.
3. Prove that each one is a subspace of the appropriate ambient space.

### Worked answer

To find the kernel, solve `T(x,y,z) = (0,0)`. That means

\[
x+y = 0, \qquad y+z = 0.
\]

So `x = -y` and `z = -y`. Let `y = t`. Then

\[
(x,y,z) = (-t,t,-t) = t(-1,1,-1).
\]

Hence

\[
\ker(T) = \operatorname{span}\{(-1,1,-1)\}.
\]

Now find the image. Evaluate `T` on the standard basis vectors:

\[
T(1,0,0) = (1,0), \qquad T(0,1,0) = (1,1), \qquad T(0,0,1) = (0,1).
\]

The vectors `(1,0)` and `(0,1)` are both in the image, so the image contains a basis of `\mathbb{R}^2`. Therefore

\[
\operatorname{im}(T) = \mathbb{R}^2.
\]

Why are these subspaces?

- The kernel is a span of one vector, so it is a subspace of `\mathbb{R}^3`.
- The image is all of `\mathbb{R}^2`, and the whole ambient space is always a subspace of itself.

If you want the closure proof explicitly for the kernel, let `u,v \in \ker(T)`. Then `T(u)=0` and `T(v)=0`, so linearity gives

\[
T(u+v) = T(u) + T(v) = 0 + 0 = 0
\]

and

\[
T(cu) = cT(u) = c \cdot 0 = 0.
\]

So the kernel is closed under addition and scalar multiplication.

### Structural sentence

The kernel measures which input directions get completely flattened by the map, while the image records every output direction the map can actually reach.

### Likely wrong move

Computing a basis for the nullspace and then never saying why kernels or images should be subspaces leaves the algebra detached from the underlying linear-map structure.


## Problem 3 — extend an independent set to a basis

Start with the independent set

\[
u_1 = (1,1,0), \qquad u_2 = (0,1,1)
\]

in `\mathbb{R}^3`.

### Prompt

1. Verify that `u_1` and `u_2` are linearly independent.
2. Add one vector to extend them to a basis of `\mathbb{R}^3`.
3. Explain why the new set spans all of `\mathbb{R}^3`.

### Worked answer

For independence, solve

\[
a(1,1,0) + b(0,1,1) = (0,0,0).
\]

This gives

\[
(a, a+b, b) = (0,0,0),
\]

so `a=0` from the first coordinate and `b=0` from the third. Thus `u_1,u_2` are independent.

Now try `u_3 = (1,0,0)`. To test whether `u_3` is already in the span of `u_1,u_2`, solve

\[
a(1,1,0) + b(0,1,1) = (1,0,0).
\]

From the first coordinate, `a=1`. From the third, `b=0`. But then the second coordinate becomes `a+b = 1`, not `0`. So `u_3` is not in the span of `u_1,u_2`.

Therefore the set

\[
\bigl\{(1,1,0),(0,1,1),(1,0,0)\bigr\}
\]

is linearly independent. Any independent set of three vectors in `\mathbb{R}^3` is automatically a basis, so this extends the original set to a basis of `\mathbb{R}^3`.

If you want to see spanning directly, solve

\[
a(1,1,0) + b(0,1,1) + c(1,0,0) = (x,y,z).
\]

This becomes

\[
a+c = x, \qquad a+b = y, \qquad b = z.
\]

So `b=z`, `a=y-z`, and `c=x-y+z`, which always exists. Hence the set spans all of `\mathbb{R}^3`.

### Structural sentence

Basis extension is not magic. It is the controlled act of adding a genuinely new direction until the space has no room left for another independent one.

### Likely wrong move

Picking a third vector because it 'looks different' without checking whether it already lies in the old span is exactly how fake basis arguments happen.


## Problem 4 — a clean diagonalization failure

Consider the matrix

\[
A = \begin{bmatrix} 1 & 1 \\ 0 & 1 \end{bmatrix}.
\]

### Prompt

1. Find the eigenvalues of `A`.
2. Find the eigenvectors.
3. Explain why `A` is not diagonalizable.
4. Say what structural failure happened.

### Worked answer

The characteristic polynomial is

\[
\det(A-\lambda I) = \det \begin{bmatrix} 1-\lambda & 1 \\ 0 & 1-\lambda \end{bmatrix} = (1-\lambda)^2.
\]

So the only eigenvalue is `\lambda = 1`, with algebraic multiplicity `2`.

Now solve `(A-I)v = 0`:

\[
A-I = \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix}.
\]

If `v = (x,y)`, then

\[
(A-I)v = (y,0),
\]

so we need `y=0`. Therefore every eigenvector has the form

\[
(x,0), \qquad x \neq 0.
\]

That means the eigenspace is only one-dimensional:

\[
E_1 = \operatorname{span}\{(1,0)\}.
\]

A `2 \times 2` matrix is diagonalizable only if it has a basis of two linearly independent eigenvectors. This matrix has only one independent eigenvector, so it is not diagonalizable.

### Structural sentence

The problem is not that the eigenvalue calculation went badly. The problem is that one repeated eigenvalue did not produce enough independent invariant directions to build an eigenbasis.

### Likely wrong move

Seeing a repeated eigenvalue and assuming diagonalization is automatic confuses algebraic multiplicity with the number of available eigenvector directions.
