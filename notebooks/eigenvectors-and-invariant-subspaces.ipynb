# Eigenvectors and invariant subspaces

Use this after `notes/linear-algebra-bridge-follow-up-packet.md`.
The point here is not to add more theorem names. It is to make diagonalization feel structural: a matrix becomes simpler when the space splits into lanes that the map already preserves.


## Setup and one sentence to keep

A subspace `W` is **invariant** for a linear map `T` if `T(W) \subseteq W`.
An eigenvector spans the smallest nontrivial example of that idea: a one-dimensional invariant subspace.

If you remember one sentence from this notebook, make it this one:

> diagonalization is a decomposition of the whole space into invariant lines.


## Problem 1 — from eigenvector to invariant line

Let

\[
A = \begin{pmatrix} 3 & 0 \\ 0 & -2 \end{pmatrix}.
\]

Show that the line `W = \operatorname{span}\{(1,0)\}` is invariant, and describe how `A` acts on `W`.

### Worked answer

Any vector in `W` has the form `(c,0)`. Then

\[
A(c,0) = (3c,0),
\]

which still lies in `W`. So `W` is invariant. On that line, `A` acts by multiplication by `3`.

### Structural sentence

This line is closed because the matrix never mixes the first coordinate into the second one.

### Likely wrong move

Saying `(1,0)` is an eigenvector but never checking what happens to an arbitrary vector `c(1,0)` misses the subspace statement.


## Problem 2 — a preserved plane is larger than an eigenline

Let

\[
A = \begin{pmatrix} 2 & 1 & 4 \\ 0 & 3 & -1 \\ 0 & 0 & 5 \end{pmatrix}.
\]

Show that `W = \operatorname{span}\{e_1,e_2\}` is invariant. Then write the matrix of the restricted map `A|_W` in the basis `(e_1,e_2)`.

### Worked answer

A general vector in `W` has the form `(x,y,0)`. Applying `A` gives

\[
A(x,y,0) = (2x+y, 3y, 0),
\]

which is still in `W`. So the plane is invariant. In the basis `(e_1,e_2)`, the restricted matrix is

\[
\begin{pmatrix} 2 & 1 \\ 0 & 3 \end{pmatrix}.
\]

### Structural sentence

The first two coordinates form a closed subsystem even before the whole matrix is fully decomposed.

### Likely wrong move

Treating `W` as invariant only because the matrix is upper triangular is too fast unless you can say what happens to `(x,y,0)`.


## Problem 3 — diagonalization as invariant-line bookkeeping

Let

\[
A = \begin{pmatrix} 4 & 0 \\ 0 & 1 \end{pmatrix}.
\]

Explain why `\mathbb{R}^2` decomposes into two invariant lines for `A`, and why that already explains the diagonal form.

### Worked answer

The lines `\operatorname{span}\{e_1\}` and `\operatorname{span}\{e_2\}` are both invariant because

\[
Ae_1 = 4e_1, \qquad Ae_2 = e_2.
\]

They span the whole space and intersect only at `0`, so

\[
\mathbb{R}^2 = \operatorname{span}\{e_1\} \oplus \operatorname{span}\{e_2\}.
\]

The matrix is diagonal because each basis vector stays in its own invariant line and only gets rescaled.

### Structural sentence

A diagonal matrix is the coordinate shadow of a space already split into independent invariant directions.

### Likely wrong move

Thinking the matrix is diagonal first and only then talking about eigenvectors reverses the real idea. The invariant splitting is the reason the diagonal form exists.


## Problem 4 — two different failure stories

Compare these matrices over `\mathbb{R}`:

\[
R = \begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix},
\qquad
J = \begin{pmatrix} 2 & 1 \\ 0 & 2 \end{pmatrix}.
\]

Why does neither matrix diagonalize over `\mathbb{R}`, and why are the reasons different?

### Worked answer

For `R`, every nonzero real vector is rotated off its own line, so there is no nontrivial real invariant line at all. The whole plane is invariant, but there is no real eigenbasis because there are no real eigenvectors.

For `J`, there is a real eigenvalue `2`, and the line `\operatorname{span}\{(1,0)\}` is invariant. But that is the only eigenline. There are not enough independent eigenvectors to form a basis, so diagonalization still fails.

### Structural sentence

Rotation fails because no real invariant lines exist. The Jordan block fails because one invariant line exists, but not enough of them.

### Likely wrong move

Saying both matrices are “not diagonalizable” without separating the geometric reason from the defective-eigenvector reason hides two very different phenomena.
