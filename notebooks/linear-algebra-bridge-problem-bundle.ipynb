## Problem 1 — span and dependence without matrix tunnel vision

Let

\[
v_1 = (1,0,1),\qquad v_2 = (0,1,1),\qquad v_3 = (1,1,2),\qquad w = (2,1,3).
\]

### Prompt

1. Decide whether `w` lies in `span(v_1, v_2)`.
2. Decide whether `v_1, v_2, v_3` are linearly independent.
3. Give the structural relation, not only the arithmetic relation.

### Worked answer

First check `span(v_1, v_2)`:

\[
a v_1 + b v_2 = a(1,0,1) + b(0,1,1) = (a,b,a+b).
\]

To get `w = (2,1,3)`, choose `a = 2` and `b = 1`.
Then the third coordinate also matches because `a+b = 3`.
So

\[
w = 2v_1 + v_2,
\]

which means `w` is in the span.

Now compare `v_3` with `v_1 + v_2`:

\[
v_1 + v_2 = (1,0,1) + (0,1,1) = (1,1,2) = v_3.
\]

So the set `\{v_1,v_2,v_3\}` is linearly dependent.
A nontrivial dependence relation is

\[
v_3 - v_1 - v_2 = 0.
\]

### Structural sentence

The real fact is not just that one equation worked.
It is that `v_1` and `v_2` already generate the plane slice `z = x + y`, and both `w` and `v_3` live inside that same span.


## Problem 2 — basis and dimension in a subspace you can describe in words

Consider the subspace

\[
S = \{(x,y,z) \in \mathbb{R}^3 : x+y+z=0\}.
\]

### Prompt

1. Show that `S` is spanned by two vectors.
2. Produce a basis for `S`.
3. Explain why the dimension is `2`.

### Worked answer

From `x+y+z=0`, we get `z = -x-y`.
So every vector in `S` has the form

\[
(x,y,-x-y) = x(1,0,-1) + y(0,1,-1).
\]

Therefore

\[
S = \operatorname{span}\{(1,0,-1),(0,1,-1)\}.
\]

To check that these two spanning vectors are independent, solve

\[
a(1,0,-1) + b(0,1,-1) = (0,0,0).
\]

This gives `(a,b,-a-b) = (0,0,0)`, hence `a=0` and `b=0`.
So the spanning set is independent and is therefore a basis.

A valid basis is

\[
\bigl\{(1,0,-1),(0,1,-1)\bigr\}.
\]

Since there are two vectors in the basis, the dimension is `2`.

### Structural sentence

The equation `x+y+z=0` cuts one degree of freedom out of `\mathbb{R}^3`, so the subspace is a plane through the origin, not a line and not all of space.


## Problem 3 — what row reduction says about the **original** columns

Let

\[
A=
\begin{bmatrix}
1 & 2 & 1 \\
2 & 4 & 0 \\
1 & 2 & 2
\end{bmatrix}.
\]

### Prompt

1. Row-reduce `A`.
2. Identify the pivot columns.
3. Say what that tells you about dependence among the **original** columns.
4. Solve `Ax=0` and interpret the free variable structurally.

### Worked answer

Start with

\[
\begin{bmatrix}
1 & 2 & 1 \\
2 & 4 & 0 \\
1 & 2 & 2
\end{bmatrix}.
\]

Apply `R_2 \leftarrow R_2 - 2R_1` and `R_3 \leftarrow R_3 - R_1`:

\[
\begin{bmatrix}
1 & 2 & 1 \\
0 & 0 & -2 \\
0 & 0 & 1
\end{bmatrix}.
\]

Then scale the third row and clean up:

\[
\begin{bmatrix}
1 & 2 & 0 \\
0 & 0 & 1 \\
0 & 0 & 0
\end{bmatrix}.
\]

So the pivot columns are columns `1` and `3` of the reduced matrix, which means columns `1` and `3` of the **original** matrix form a basis for the column space.
The second original column is dependent on them.
In fact,

\[
\begin{bmatrix}2\\4\\2\end{bmatrix} = 2\begin{bmatrix}1\\2\\1\end{bmatrix}.
\]

To solve `Ax=0`, use the reduced system:

\[
x_1 + 2x_2 = 0,\qquad x_3 = 0.
\]

Let `x_2 = t`.
Then `x_1 = -2t`, so

\[
x = t(-2,1,0).
\]

### Structural sentence

The free variable records one redundancy among the columns: one column direction was not buying a new direction in the output space.


## Problem 4 — a matrix is a map before it is an array

Define

\[
T : \mathbb{R}^2 \to \mathbb{R}^2,\qquad T(x,y) = (x+2y,\, y-x).
\]

### Prompt

1. Find the standard matrix of `T`.
2. Compute `\ker(T)` and `\operatorname{im}(T)`.
3. Decide whether `T` is invertible.
4. Describe the map in words before talking about determinants.

### Worked answer

Apply `T` to the standard basis:

\[
T(1,0) = (1,-1),\qquad T(0,1) = (2,1).
\]

So the standard matrix is

\[
[T]_{
\text{std}} =
\begin{bmatrix}
1 & 2 \\
-1 & 1
\end{bmatrix}.
\]

For the kernel, solve

\[
\begin{bmatrix}
1 & 2 \\
-1 & 1
\end{bmatrix}
\begin{bmatrix}x\\y\end{bmatrix}
=
\begin{bmatrix}0\\0\end{bmatrix}.
\]

This gives

\[
x+2y=0,\qquad -x+y=0.
\]

From the second equation, `y=x`.
Substitute into the first: `x+2x=3x=0`, so `x=0` and `y=0`.
Hence

\[
\ker(T)=\{(0,0)\}.
\]

Because the kernel is trivial and the domain/codomain both have dimension `2`, the map is invertible.
Its image is all of `\mathbb{R}^2`.

### Structural sentence

This map mixes the coordinates rather than killing one direction. Nothing collapses to zero except the zero vector, so the plane gets reshuffled, not flattened.


## Problem 5 — change of basis as a simplification story

Let

\[
A=
\begin{bmatrix}
3 & 1 \\
1 & 3
\end{bmatrix},
\qquad b_1=(1,1),\qquad b_2=(1,-1).
\]

### Prompt

1. Show that `b_1` and `b_2` form a basis of `\mathbb{R}^2`.
2. Compute `A b_1` and `A b_2`.
3. Find the matrix of the same linear map in the basis `B=(b_1,b_2)`.
4. Explain what this basis reveals.

### Worked answer

The vectors `(1,1)` and `(1,-1)` are independent because neither is a scalar multiple of the other, so they form a basis.

Now apply `A`:

\[
A b_1 =
\begin{bmatrix}
3 & 1 \\
1 & 3
\end{bmatrix}
\begin{bmatrix}1\\1\end{bmatrix}
=
\begin{bmatrix}4\\4\end{bmatrix}
=4b_1,
\]

and

\[
A b_2 =
\begin{bmatrix}
3 & 1 \\
1 & 3
\end{bmatrix}
\begin{bmatrix}1\\-1\end{bmatrix}
=
\begin{bmatrix}2\\-2\end{bmatrix}
=2b_2.
\]

So in the basis `B`, the matrix is already diagonal:

\[
[A]_B =
\begin{bmatrix}
4 & 0 \\
0 & 2
\end{bmatrix}.
\]

### Structural sentence

In the standard basis the map looks like coordinate mixing.
In the basis `B`, the same map reveals its real behavior immediately: it stretches one invariant direction by `4` and the other by `2`.
That is the first honest reason eigenvectors matter.


## Retrospective audit after each problem

After finishing a problem, write short answers to these:

1. What was the structural object here: span, subspace, column space, kernel, or basis change?
2. Which line of algebra actually carried the meaning?
3. What would have gone wrong if I treated this as pure symbol pushing?
4. Could I explain the result out loud without saying "because the algorithm says so"?
5. If I got stuck, was the problem proof load, notation load, or concept load?

That last question matters.
The bridge fails when every stall gets mislabeled as "I just need more practice" even when the real issue is that the definitions still do not have stable meaning.


## Adversarial check

Do **not** conclude that the bridge is complete just because these examples feel readable.

The bridge is working only if you can now do things like:

- invent a small dependence relation on your own,
- explain why pivot columns refer back to the original columns,
- move between a matrix picture and a map picture without freezing,
- and say what a clever basis is buying you before you compute it.

If those still collapse after one day away from the notebook, stay in preview mode for another week.
That is not failure.
That is exactly what the bridge is for.
